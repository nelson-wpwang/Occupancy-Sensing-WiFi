```python
fig = plt.figure(figsize=(16, 4), dpi=80)
ax1 = fig.add_subplot(111)

len_plt=200

ax1.scatter(x=ftm0_data['index'][1:len_plt], y=ftm0_data['ch2'][1:len_plt], s=10, c='b', marker="s", label='ch2')
ax1.scatter(x=ftm0_data['index'][1:len_plt], y=ftm0_data['ch9'][1:len_plt], s=10, c='r', marker="o", label='ch9')
ax1.scatter(x=ftm0_data['index'][1:len_plt], y=ftm0_data['ch3'][1:len_plt], s=10, c='g', marker="o", label='ch3')
ax1.scatter(x=ftm0_data['index'][1:len_plt], y=ftm0_data['ch8'][1:len_plt], s=10, c='y', marker="s", label='ch8')


plt.legend(loc='upper right')
plt.xlabel('Sample')
plt.ylabel('FTM value')
plt.title('No occupant')
plt.show()
```
